###  What are the first 10 customers in the database?

```python
%%sql
SELECT *
FROM CUSTOMERS
LIMIT 10;
```

### How many customers does the business have?

```python
%%sql
SELECT COUNT(*) AS 'Number of Customers'
FROM CUSTOMERS;
```

### Which countries have the most customers?

```python
%%sql
SELECT country, COUNT(*) AS 'Number of Customers'
FROM CUSTOMERS
GROUP BY country
ORDER BY 'Number of Customers' DESC;
```

### What are the top 10 invoices by total amount?
  These customers may be important for retention or premium offers.

```python
%%sql
SELECT
    InvoiceId,
    CustomerId,
    InvoiceDate,
    BillingCountry,
    Total
FROM invoices
ORDER BY Total DESC
LIMIT 10;
```

### What is the total revenue of the business?
This is the main business metric and will be used later for country, customer, and product-level analysis.

```python
%%sql
SELECT SUM(Total) AS 'Total Revenue'
FROM invoices;
```

### Which countries generate the most revenue?

```python
%%sql
SELECT
    BillingCountry,
    '$' || ROUND(SUM(Total),2) AS total_revenue
FROM invoices
GROUP BY BillingCountry
ORDER BY total_revenue DESC;
```

###Who are the top 10 customers by total spending?
These are the highest-spending customers. The business could target them with loyalty offers, early access, or personalised recommendations.

```python
%%sql
SELECT
    c.CustomerId,
    c.FirstName,
    c.LastName,
    '$' || ROUND(SUM(i.Total),2) AS total_spending
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
GROUP BY c.CustomerId
ORDER BY total_spending DESC
```

### Which music genres generate the most revenue?

```python
%%sql
SELECT
    g.Name AS genre,
    ROUND(SUM(ii.UnitPrice * ii.Quantity),2) AS total_revenue
FROM invoice_items ii
JOIN tracks t
    ON ii.TrackId = t.TrackId
JOIN genres g
    ON t.GenreId = g.GenreId
GROUP BY g.Name
ORDER BY total_revenue DESC;
```

### Who are the top 10 artists by revenue?
These artists generate the most revenue. The company could feature them more prominently, create artist-based campaigns, or recommend similar artists to customers.

```python
%%sql
SELECT
    artists.Name AS artist,
    ROUND(SUM(ii.UnitPrice * ii.Quantity),2) AS total_revenue
FROM invoice_items ii
JOIN artists
ON artists.ArtistId = albums.ArtistId
JOIN albums
ON albums.AlbumId = t.AlbumId
JOIN tracks t
ON t.TrackId = ii.TrackId
GROUP BY artists.Name
ORDER BY total_revenue DESC
LIMIT 10;
```
